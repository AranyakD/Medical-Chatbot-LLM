```python
import pandas as pd
import re 
```

```python
df = pd.read_csv('medquad.csv')
print(df.shape)
df.head()
```

```python
df = df.drop_duplicates()
df 
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['question', 'answer'])
df
```

```python
# Striping whitespace
df.loc[:, "question"] = df["question"].str.strip()
df.loc[:, "answer"]   = df["answer"].str.strip()

# Filtering empty rows
df = df[(df["question"] != "") & (df["answer"] != "")]
```

```python
df = df[["question", "answer"]]
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"[^a-z0-9.,? ]", "", text)
    return text.strip()

df["question"] = df["question"].apply(clean_text)
df["answer"] = df["answer"].apply(clean_text)
df
```

```python
DISCLAIMER = (
    " this information is for general guidance only and does not replace professional medical advice."
)

df["answer"] = df["answer"] + DISCLAIMER
```

```python
df.to_csv("medquad_cleaned.csv", index=False)
df
```

```python
import pandas as pd

df = pd.read_csv("medquad_cleaned.csv")
print(df.shape) 
```

```python
from sklearn.model_selection import train_test_split

# First split = training (80%) and temporary (20%)
train_df, temp_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42
)

# Second split = validation (10%) and test (10%)
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    random_state=42
)

print("Training set:", train_df.shape)
print("Validation set:", val_df.shape)
print("Test set:", test_df.shape)
```

```python
# converting from pandas dataframe to huggingface dataset
from datasets import Dataset
train_dataset = Dataset.from_pandas(train_df)
val_dataset   = Dataset.from_pandas(val_df)
test_dataset  = Dataset.from_pandas(test_df)
train_dataset = train_dataset.remove_columns(["__index_level_0__"])
val_dataset   = val_dataset.remove_columns(["__index_level_0__"])
test_dataset  = test_dataset.remove_columns(["__index_level_0__"])
train_dataset[0]

```

```python
from transformers import AutoTokenizer
model_name = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)

```

```python
def tokenize_function(batch):
    # Tokenize input 
    model_inputs = tokenizer(
        batch["question"],
        padding="max_length",
        truncation=True,
        max_length=128
    )

    # Tokenize output 
    labels = tokenizer(
        text_target=batch["answer"],
        padding="max_length",
        truncation=True,
        max_length=200
        )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset   = val_dataset.map(tokenize_function, batched=True)
```

```python
train_dataset = train_dataset.remove_columns(
    ["question", "answer"]
)

val_dataset = val_dataset.remove_columns(
    ["question", "answer"]
)


```
